```python
import pandas     as pd
import numpy      as np
#import matplotlib.pyplot as plt
from matplotlib import pyplot, pylab
```

```python
data1 = pd.read_csv('progress100-128.csv')
data2 = pd.read_csv('progress10k-128.csv')
data3 = pd.read_csv('progress10k-256.csv')
data4 = pd.read_csv('progress10k-512.csv')
data5 = pd.read_csv('progress10k-768.csv')
data6 = pd.read_csv('progress10k-896.csv')
# data6 = pd.read_csv('progress10k-1024-100.csv')

data4.head()
```

```python
pyplot.figure(figsize=(15,5))
for data in [data1, data2, data3, data4, data5, data6]:
    pyplot.plot(data['epoch'], data['test/mean_Q'], label='sample')
pylab.xlabel('Epoch')
pylab.ylabel('Mean Q value')
#pyplot.fill_between(color=color)
#pyplot.xlim(0,6)
pyplot.title('Training results comparison for Fetch-PickAndPlace')
L=pyplot.legend()

L.get_texts()[0].set_text('100-128')
L.get_texts()[1].set_text('10k-128')
L.get_texts()[2].set_text('10k-256')
L.get_texts()[3].set_text('10k-512')
L.get_texts()[4].set_text('10k-768')
L.get_texts()[5].set_text('10k-896')
pyplot.savefig('100_10k-128-896QValue.png')
```
